```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="Set2")

# Load data from CSV files
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")


# Quick check of data
print(matches.head())
print(deliveries.head())



# Clean column names
matches.columns = matches.columns.str.strip()
deliveries.columns = deliveries.columns.str.strip()

# Function to add value annotations to bar plots
def annotate_bars(ax, orientation='h', space=0.5):
    for p in ax.patches:
        if orientation == 'h':
            ax.text(p.get_width() + space, p.get_y() + p.get_height() / 2,
                    int(p.get_width()), va='center', fontsize=10)
        else:
            ax.text(p.get_x() + p.get_width() / 2, p.get_height() + space,
                    int(p.get_height()), ha='center', fontsize=10)

# 1. Total Wins by Each Team
if 'WinningTeam' in matches.columns:
    team_wins = matches['WinningTeam'].value_counts().sort_values()
    plt.figure(figsize=(12, 7))
    ax = sns.barplot(x=team_wins.values, y=team_wins.index, palette='Spectral')
    annotate_bars(ax)
    plt.title("🏆 Most Successful IPL Teams (By Wins)", fontsize=18, fontweight='bold')
    plt.xlabel("Total Wins", fontsize=14)
    plt.ylabel("Team", fontsize=14)
    plt.tight_layout()
    plt.show()

# 2. Toss Decision Analysis
if 'TossDecision' in matches.columns:
    toss_decision = matches['TossDecision'].value_counts()
    plt.figure(figsize=(6, 4))
    ax = sns.barplot(x=toss_decision.index, y=toss_decision.values, palette='Set3')
    annotate_bars(ax, orientation='v', space=1)
    plt.title("🪙 Toss Decision Trends", fontsize=16, fontweight='bold')
    plt.xlabel("Decision", fontsize=12)
    plt.ylabel("Number of Matches", fontsize=12)
    plt.tight_layout()
    plt.show()

# 3. Top 10 Run Scorers
if {'batter', 'batsman_run'}.issubset(deliveries.columns):
    top_scorers = deliveries.groupby('batter')['batsman_run'].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 7))
    ax = sns.barplot(x=top_scorers.values, y=top_scorers.index, palette='magma')
    annotate_bars(ax)
    plt.title("🏏 Top 10 Run Scorers in IPL", fontsize=18, fontweight='bold')
    plt.xlabel("Total Runs", fontsize=14)
    plt.ylabel("Batter", fontsize=14)
    plt.tight_layout()
    plt.show()

# 4. Top 10 Player of the Match Awards
if 'Player_of_Match' in matches.columns:
    top_players = matches['Player_of_Match'].value_counts().head(10)
    plt.figure(figsize=(12, 7))
    ax = sns.barplot(x=top_players.values, y=top_players.index, palette='coolwarm')
    annotate_bars(ax)
    plt.title("⭐ Top 10 Player of the Match Award Winners", fontsize=18, fontweight='bold')
    plt.xlabel("Awards", fontsize=14)
    plt.ylabel("Player", fontsize=14)
    plt.tight_layout()
    plt.show()

# 5. Matches Played Per Season
if 'Season' in matches.columns:
    matches_per_season = matches['Season'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=matches_per_season.index, y=matches_per_season.values, 
                 marker='o', linewidth=2.5, color='navy')
    plt.title("📅 IPL Matches Played Per Season", fontsize=18, fontweight='bold')
    plt.xlabel("Season", fontsize=14)
    plt.ylabel("Number of Matches", fontsize=14)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Toss decision win percentage
if {'TossDecision', 'TossWinner', 'WinningTeam'}.issubset(matches.columns):
    toss_wins = matches[matches['TossWinner'] == matches['WinningTeam']]['TossDecision'].value_counts()
    toss_total = matches['TossDecision'].value_counts()
    toss_win_pct = (toss_wins / toss_total * 100).fillna(0)

    plt.figure(figsize=(6, 6))
    plt.pie(toss_win_pct, labels=toss_win_pct.index, autopct='%1.1f%%',
            colors=sns.color_palette("Set2"), startangle=140, wedgeprops={'edgecolor': 'black'})
    plt.title("🥧 Win Percentage by Toss Decision", fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Head-to-head matrix of wins
if {'Team1', 'Team2', 'WinningTeam'}.issubset(matches.columns):
    head_to_head = pd.crosstab(index=matches['WinningTeam'], columns=matches['Team2'])
    plt.figure(figsize=(14, 10))
    sns.heatmap(head_to_head, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5, linecolor='white')
    plt.title("🧮 Team vs Team Win Count Matrix", fontsize=18, fontweight='bold')
    plt.xlabel("Defeated Team", fontsize=14)
    plt.ylabel("Winning Team", fontsize=14)
    plt.tight_layout()
    plt.show()
```
